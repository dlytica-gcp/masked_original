```python
import pandas as pd
charge_revol_transaction=pd.read_excel("charge_revol_transaction.xlsx")
charge_revol_transaction.head()
```

```python
charge_revol_transaction['crt_mer_ense'].value_counts()
```

```python
def rename_merchant(name):
    if isinstance(name, str):
        parts = name.split()
        if len(parts) == 0:
            return "Dlytica"
        elif parts[0].upper() == 'BBSM':
            return 'Bigmart ' + ' '.join(parts[1:])
        else:
            return 'Dlytica Inc ' + ' '.join(parts[1:])
    return "Dlytica Inc"

charge_revol_transaction['crt_mer_ense'] = charge_revol_transaction['crt_mer_ense'].apply(rename_merchant)
```
