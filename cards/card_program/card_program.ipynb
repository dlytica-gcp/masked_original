```python
import pandas as pd
import random
df_card_program=pd.read_excel("card_program.xlsx")
df_card_program.head()
```

```python
companies = ['Dlytica','Fusefire']

def replace_with(val):
    if pd.isna(val):
        return val
    words = val.split()
    last_two = ' '.join(words[-2:]) if len(words) >= 2 else val
    company = random.choice(companies)
    return f'{company} {last_two}'

df_card_program['cpr_labe'] = df_card_program['cpr_labe'].apply(replace_with)


df_card_program[['cpr_labe']].sample(10)
```
