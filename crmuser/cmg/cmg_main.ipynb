```python
import pandas as pd
```

```python
df_cmg= pd.read_excel("cmg.xlsx")
df_cmg.head()
```

```python
df_cmg = df_cmg.drop(['_airbyte_ab_id', '_airbyte_emitted_at',"source_file_path","_airbyte_additional_properties"], axis=1)

df_cmg.head()
 
```

```python
df_account=pd.read_csv(r"..\account\account_masked.csv")
df_account.head()
```

```python
df_result = pd.merge(df_cmg, df_account, left_on='cif_id',right_on='groupid',how='left',suffixes=('_p', '_a'))

df_result['cif_id_a'].head(5)
 
```

```python
df_cmg['cif_id']=df_result['cif_id_a'].values
df_cmg['email']=df_result['preferredemail'].values
df_cmg['phone']=df_result['preferredphone'].values
df_cmg['corp_id']=df_result['corp_id_a'].values
df_cmg['cust_sex']=df_result['gender'].values
df_cmg['cust_id']=df_result['core_cust_id'].values
df_cmg['cust_name']=df_result['name'].v
df_cmg['pan_gir_num']=df_result['pan']
df_cmg['ichg_userid']=df_result['bomodifiedby']
df_cmg['rcre_user_id']=df_result['bocreatedby']
df_cmg['date_of_birth']=df_result['cust_dob']
df_cmg['cust_hlth_code']=df_result['cust_health_ref_code']
df_cmg['cust_first_name']=df_result['cust_first_name_a']
df_cmg['cust_middle_name']=df_result['cust_middle_name_a']
df_cmg['cust_last_name']=df_result['cust_last_name_a']
df_cmg['cust_occp_code']=df_result['occupation']
df_cmg['nat_id_card_num']=df_result['nat_id_card_num_a']
df_cmg['cust_introd_name']=df_result['introducername']
df_cmg['cust_introd_cust_id']=df_result['introducerid']
df_cmg['bank_id']=df_result['bank_id_a']
```
