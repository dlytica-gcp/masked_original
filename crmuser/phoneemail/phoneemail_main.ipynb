```python
import pandas as pd
```

```python
df_phoneemail = pd.read_excel("phoneemail.xlsx")
df_phoneemail.head()
```

```python
df_phoneemail = df_phoneemail.drop(['_airbyte_ab_id', '_airbyte_emitted_at',"source_file_path","_airbyte_additional_properties"], axis=1)
df_phoneemail.head()
```

```python
print(len(df_phoneemail.columns))
```

```python
print(len(df_phoneemail))
```

```python
df_acc = pd.read_csv("../account/account_masked.csv")
df_acc.head()
```

```python
df_acc.columns
```

```python
result_df = pd.merge(df_phoneemail, df_acc, how='left', on='orgkey',suffixes=('_p', '_a'))
result_df[["orgkey","cif_id","phonenocitycode"]].head(80)
```

```python
print(len(result_df))
```

```python
df_phoneemail["orgkey"] = result_df["cif_id"].values
df_phoneemail["url"]= result_df["url_a"].values
df_phoneemail["email"]= result_df["email_a"].values

df_phoneemail["phoneno"]= result_df["preferredphone"].values
df_phoneemail["extensionid"]= result_df["extensionid_a"].values
df_phoneemail["accountid"]= result_df["accountid_a"].values
df_phoneemail["phonenolocalcode"]= result_df["preferredphone"].values
df_phoneemail.head()
```

```python
print(len(df_phoneemail))
```

```python
df_phoneemail.to_csv('phoneemail_masked.csv', index=False) 
```
