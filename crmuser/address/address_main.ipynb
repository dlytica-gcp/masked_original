```python
import pandas as pd
```

```python
df_address = pd.read_excel("address.xlsx")
df_address.head()
```

```python
df_address = df_address.drop(['_airbyte_ab_id', '_airbyte_emitted_at',"source_file_path"], axis=1)
df_address.head()
```

```python
print(len(df_address))
```

```python
df_acc = pd.read_csv("../account/account_masked.csv")
df_acc.head()
```

```python
print(len(df_acc.columns))
```

```python
result_df = pd.merge(df_address, df_acc, how='left', on='orgkey',suffixes=('_add', '_acc'))
result_df.head()
```

```python
df_address["orgkey"] = result_df["cif_id"].values
df_address["name"] = result_df["name_acc"].values
df_address["state"] = result_df["state_acc"].values
df_address["bank_id"] = result_df["bank_id_acc"].values
df_address["address_line1"] = result_df["address_line1_acc"].values
df_address["address_line2"] = result_df["address_line2_acc"].values
df_address["address_line3"] = result_df["address_line3_acc"].values
df_address["city"] = result_df["city_acc"].values
df_address["city"] = result_df["city_acc"].values
df_address["swift_name"] = result_df["name_acc"].values
df_address.head()
```

```python
df_address.to_csv('address_masked.csv', index=False) 
```
