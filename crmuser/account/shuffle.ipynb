```python
import numpy as np 
import pandas as pd
df_male = pd.read_csv('male.csv')
nepali_surnames = [
    'Thapa', 'Poudel', 'Acharya', 'Adhikari', 'Bhandari', 
    'Bhattarai', 'Dahal', 'Gurung', 'Karki', 'Khadka',
    'Lama', 'Limbu', 'Magar', 'Maharjan', 'Neupane',
    'Pandey', 'Rai', 'Rana', 'Regmi', 'Sharma',
    'Shrestha', 'Tamang', 'Thakuri', 'Timalsina', 'Yadav'
]
nepali_male_middle_names = [
    "Bahadur", "Kumar", "Prasad", "Ram", "Lal",
    "Hari", "Shankar", "Maan", "Dev", "Bhakta",
    "Krishna", "Gopal", "Narayan", "Govinda", "Bishnu",
    "Bikram", "Biren", "Bishal", "Bibek", "Bipin"]


df_male['cust_middle_name'] = np.random.choice(nepali_male_middle_names, size=len(df_male))
df_male['cust_last_name'] = np.random.choice(nepali_surnames, size=len(df_male))
df_male_final = df_male.sample(n=72, random_state=42)  # `axis=1` for columns

df_male_final["gender"] = "M"
df_male_final.head(5)
# print(len(df_male_final))
```

```python
import numpy as np 
import pandas as pd
df_female = pd.read_csv('female.csv')
nepali_surnames = [
    'Thapa', 'Poudel', 'Acharya', 'Adhikari', 'Bhandari', 
    'Bhattarai', 'Dahal', 'Gurung', 'Karki', 'Khadka',
    'Lama', 'Limbu', 'Magar', 'Maharjan', 'Neupane',
    'Pandey', 'Rai', 'Rana', 'Regmi', 'Sharma',
    'Shrestha', 'Tamang', 'Thakuri', 'Timalsina', 'Yadav'
]
nepali_female_middle_names = [
    "Maya", "Devi", "Kumari", "Laxmi", "Saraswati"]



df_female['cust_middle_name'] = np.random.choice(nepali_female_middle_names, size=len(df_female))
df_female['cust_last_name'] = np.random.choice(nepali_surnames, size=len(df_female))

df_female_final = df_female.sample(n=53, random_state=42)

df_female_final["gender"] = "F"
df_female_final.head(15)
```

```python
result = pd.concat([df_male_final,df_female_final])
shuffled = result.sample(frac=1, random_state=42).reset_index(drop=True)
shuffled["cust_type"] = "INDIVIDUAL"
print(shuffled)
```

```python
shuffled.to_csv('shuffled_data.csv', index=False) 
```
