```python
import pandas as pd
df_acct_nom= pd.read_csv("acct_nomination_table.csv")
df_acct_nom.head()
```

```python
df_acct_nom[['acid','cif_id']]=df_result[['acid_t','cif_id_t']]
df_acct_nom.head()
```
