```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("general_acct_mast_table.xlsx")
df.head()
```

```python
df = df.drop_duplicates(subset=['acid'])
print(len(df))
```

```python
df = df.drop(['_airbyte_ab_id', '_airbyte_emitted_at',"source_file_path","_airbyte_additional_properties"], axis=1)
df.head()
```

```python
def generate_acid(data):
    return "AC" + str(np.random.randint(10000000, 99999999))

df["acid_t"] = df["acid"].apply(generate_acid)
df.head()

```

```python
import numpy as np
import pandas as pd

def generate_foracid():
    random_digit =  str(np.random.randint(1000000000000000, 9999999999999999,dtype='int64'))
    return "FA"+random_digit

df["foracid_t"] = [generate_foracid() for _ in range(len(df))]

df["acct_num"] = df["foracid_t"].str[4:16]
df.head()
```

```python
print(len(df.columns))
```

```python
selected_columns = [
    "acid",
    "acid_t",
    "cif_id",
    "foracid",
    "foracid_t",
    "acct_num",
    "acct_name",
    "hashed_no",
    "phone_num",
    "acct_short_name",
    "last_tran_id_cr",
    "last_tran_id_dr"
]
remaining_columns = [col for col in df.columns if col not in selected_columns]
```

```python
df = df[selected_columns+remaining_columns]
df.head()
```

```python
print(len(df.columns))
```

```python
import hashlib

def generate_hash(data):
    if not isinstance(data, str):
        data = str(data)
    return hashlib.sha1(data.encode('utf-8')).hexdigest().upper()

df["hashed_no"] = df["foracid"].apply(generate_hash)
```
