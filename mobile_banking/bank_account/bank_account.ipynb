```python
df_cif_mask=pd.read_csv(r"..\..\crmuser\account\cif_masked.csv")
df_cif_mask.head()
```

```python
df_acidfor_mask=pd.read_csv(r"..\..\tbaadm\general_acct_mast_table\acid_foracid_masked.csv")
df_acidfor_mask.head()
```

```python
result = pd.merge(left=df_bankxp_bank_acc, right=df_cif_mask,how='left',left_on='account_owner_cbs_id',right_on='orgkey',suffixes=('_p', '_a'))

result
```

```python
df_acidfor_mask = df_acidfor_mask.drop_duplicates(subset='cif_id')
```

```python
print(result.shape)
print(df_bankxp_bank_acc.shape)
```

```python
dresult = pd.merge(result,df_acidfor_mask,how='left',on='cif_id',suffixes=('_p', '_a'))

dresult.sample(20)
```

```python
print(dresult.shape)
print(df_bankxp_bank_acc.shape)
```

```python
dresult.columns
```

```python
df_bankxp_bank_acc['account_number']=dresult['foracid_t']
df_bankxp_bank_acc['account_holder_name']=dresult['acct_name']
df_bankxp_bank_acc['account_owner_cbs_id']=dresult['cif_id']

df_bankxp_bank_acc['account_owner_cbs_id']
```

```python
df_bankxp_bank_acc = df_bankxp_bank_acc[df_bankxp_bank_acc['account_owner_cbs_id'].notna()]
```
